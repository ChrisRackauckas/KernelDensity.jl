````julia
using PyPlot
using Distributions
using KernelDensity
````

If you load KernelDensity before PyPlot, you will need to manually call
```
KernelDensity.PyPlot_init()
```
before the plotting functions (*e.g.* `plot` and `contour`) will function correctly.

````julia
X = rand(Normal(),10);
````

````julia
k = kde(X)
````

````julia
plot(k)
````

````julia
plot(kde(X;kernel=Uniform))
````
