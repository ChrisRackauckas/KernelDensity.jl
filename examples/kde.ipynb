````julia
using Winston
using Distributions
using KernelDensity
````

If you load KernelDensity before Winston, you will need to manually call
```
KernelDensity.Winston_init()
```
before the plotting functions (`plot` and `imagesc`) will function.

````julia
X = rand(Normal(),10);
````

````julia
k = kde(X)
````

````julia
plot(k)
````

````julia
plot(kde(X;kernel=Uniform))
````
